```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline # jupyternootbook内に表示させる指定
```

```python
np.random.seed(1) # seedを入れているので乱数が固定される
x = np.arange(10) # 連番を生成
y = np.random.rand(10)

plt.plot(x,y)
plt.show()
```

```python
def f2(x,w):
    return(x-w)*x*(x+2)

x = np.linspace(-3,3,100)

plt.plot(x,f2(x,2), color='black', label='$w=2')
plt.plot(x, f2(x,1), color='cornflowerblue',
        label='$w=1')
plt.legend(loc="upper left")
plt.ylim(-15,15)
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```
