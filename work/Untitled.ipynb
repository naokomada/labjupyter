```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
np.random.seed(1) # seedを入れているので乱数が固定される
x = np.arange(10) # 連番を生成
y = np.random.rand(10)

plt.plot(x,y)

plt.plot(x,y, color='black', label='$w=2')
plt.legend(loc="upper left")
plt.ylim(-15,15)
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```
